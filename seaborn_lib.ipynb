```python
import seaborn as sns
```

```python
df = sns.load_dataset("flights")
```

```python
# flights name id dependent features
# Year , month , Passengers field is independent features
df.head()
```

```python
df
```

```python
df.dtypes
```

Correlation with Heatmap

```python
df.corr()
```

```python
# heatmap use for visulization
sns.heatmap(df.corr())
```

Univariate Analysis

```python
sns.jointplot(x='year',y='passengers',data=df,kind='hex')
```

```python
sns.jointplot(x='year',y='passengers',data=df,kind='reg')
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df,hue='month')
```

```python
df['month'].value_counts()
```

```python
sns.distplot(df['passengers'],kde = False , bins=10)
```

Count Plot

```python
sns.countplot('month',data=df)
```

```python
sns.countplot(y='month',data=df)
```

```python
#in a Bar plot you need to specify value of x and y axis
sns.barplot(x='month',y='passengers',data=df)
```

Box Plot

```python
sns.boxplot('month','passengers',data=df)
```

```python
# categorize Data (hue values are compared value)

sns.boxplot(x='year',y='passengers',hue='month',data=df)
```

Violin Plot

```python
sns.violinplot(x='passengers',y='month',data=df,palette='rainbow')
```
